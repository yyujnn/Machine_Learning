### 이미지 처리
1 픽셀 기반 처리 : 이미지를 픽셀 단위로 처리  
2. 블록 기반 처리 : 이미지를 블록 (픽셀을 그룹화 한 것 ) 단위로 처리하는 방법  
3. 주파수 기반 처리 : 이미지를 주파수 영역으로 변환하여 처리하는 방법  
- 픽셀 (pixel : picture element)
    - 이미지 구성 최소 단위
- 주파수 (frequency)
    - 이미지 픽셀값들을 픽셀값의 변화분으로 변환한 것
    - 이미지에서의 주파수 : 픽셀의 변화량
    - 고주파 성분 : 픽셀의 변화가 큰 부분(에지)
    - 저주파 성분 : 픽셀의 변화가 작은 부분

### 픽셀 값 변경

```python
img = cv2.imread('cocacola.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for x in range(200, 210):
    for y in range(100, 110):
        img[x, y] = [0, 0, 255] # 픽셀값을 파란색으로 변경

plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

### ROI에 도형 그리기

```python
# 직선 그리기 - line
# line(img, (시작점 좌표),(끝점 좌표), 색상, 선 두께)
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = cv2.line(img, (420,460), (470,460), (0,255,255) , 20)

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

```python
# 원 그리기 - circle
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# circle(이미지, 중심좌표, 반지름, 색상, 선두께)
img = cv2.circle(img, (170,280), 50, (0,255,255),5)
img = cv2.circle(img, (400,280), 50, (0,255,255),-1)

# img [40:360 , 10:485]

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

```python
# 사각형 그리기 - rectangle
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# rectangle(이미지, 좌상단좌표, 우하단좌표, 색상, 선두께)
img = cv2.rectangle(img, (100,100), (350,350), (255,255,255),5)

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

```python
# 텍스트 출력
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 사용할 폰트 지정
font = cv2.FONT_HERSHEY_SIMPLEX

# putText(이미지, 출력내용, 출력시작좌표, 폰트, 크기, 색상, 굵기)
img = cv2.putText(img, 'Text', (200,100), font, 1, (180,0,255), 3)

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

##### 한글 출력하기

```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
```

```python
# 텍스트 출력
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 사용할 폰트와 크기를 가져옴
font = ImageFont.truetype('fonts/gulim.ttc',34)
# array 이미지를 PIL이미지로 변환
image_pil = Image.fromarray(img)
# PIL이미지를 그림
draw = ImageDraw.Draw(image_pil)
# 이미지에 텍스트 그리기
draw.text((360,180), '짱구', font=font, fill= (180,0,255,0)) #(r,g,b,alpha)

# PIL 이미지를 numpy 배열로 변환
img = np.array(image_pil)

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```
