```python
import cv2
```

##### cv2를 활용하여 이미지 출력

```python
# 이미지 가져오기
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR) # 3차원

# 이미지 출력
cv2.imshow('jjang', img)

# 창 닫기
cv2.waitKey(0) # 키보드의 입력을 기다리는 함수(시간)
# 1000 -> 1초
# 0 : 계속 기다림
cv2.destroyAllWindows() # 생성된 윈도우 창을 모두 닫는다
```

- OpenCV 색상 공간을 BGR(Blue, Green, Red)로 사용
- Python 이미지들은 RGB(Red, Green, Blue)로 사용

- Blue, Red의 색상 공간이 다름-> 바꿔주자!

```python
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)

# 색상 공간 변환 (BGR -> RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

##### 직접 Red 채널과 Blue 채널 바꾸기

```python
img = cv2.imread('jjang.jpg', cv2.IMREAD_COLOR)

b, g, r = cv2.split(img) # img를 b, g, r 변수에 분리
img2 = cv2.merge([r,g,b]) # 파이썬 색상공간 순서로 합치기

# a, b = ['a', 'b']
# print(a)
# print(b)

plt.xticks([])
plt.yticks([])
plt.imshow(img2)
```
